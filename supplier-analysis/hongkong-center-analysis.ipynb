```python
hongkong_df['scandate'] = pd.to_datetime(hongkong_df['scandate'], format='%Y%m%d%H%M%S', errors="coerce")

hongkong_df['month_year'] = pd.to_datetime(hongkong_df['scandate']).dt.to_period('M')

hongkong_df = hongkong_df.fillna('unknown')
```

```python
hongkong_df
hk_scans_per_donor = hongkong_df.groupby(['donor', 'month_year']).count().reset_index()
```

```python
alt.Chart(hk_scans_per_donor.astype('str')).mark_bar().encode(
    x='yearmonth(month_year):T',
    y='scandate:Q',
    color= 'donor:N',
    tooltip = ['donor:N', 'month_year:T', 'scandate:Q']
).interactive()
```

```python
hk_scans_per_donor.loc[hk_scans_per_donor['donor'] != 'unknown']

known_donors = alt.Chart(hk_scans_per_donor.loc[hk_scans_per_donor['donor'] != 'unknown'].astype('str')).mark_bar().encode(
    x=alt.X('yearmonth(month_year):T', axis=alt.Axis(title="Month Scanned")),
    y=alt.Y('scandate:Q', axis=alt.Axis(title="Books Scanned")),
    color= 'donor:N',
    tooltip = ['donor:N', 'month_year:T', 'scandate:Q']
).interactive()

total_donors = alt.Chart(hk_scans_per_donor.astype('str')).mark_bar(color='black').encode(
    x='yearmonth(month_year):T',
    y='scandate:Q',
).interactive()

total_donors + known_donors
```
