```python
supplier_ids = bols['supplier_location_id'].unique().tolist()
company_ids = bols['company_location_id'].unique().tolist()
```

```python
company_ids
```

```python
# creating a key for locations containing ids, lats, longs
supplier_ids
```

```python
# id,name,city,country,latitude,longitude
bols.keys()
```

```python
location_key = pd.DataFrame(columns=['id', 'name', 'address','lat', 'lon'])
```

```python


for entry in supplier_ids: 
    name = bols.loc[bols['supplier_location_id'] == entry]['supplier'].tolist()[0]
    address = bols.loc[bols['supplier_location_id'] == entry]['supplier_address'].tolist()[0]
    lat = bols.loc[bols['supplier_location_id'] == entry]['supplier_lat'].tolist()[0]
    lon = bols.loc[bols['supplier_location_id'] == entry]['supplier_lon'].tolist()[0]
    location_key  = location_key.append({'id': entry, 
                        'name': name,
                        'address': address,
                        'lat': lat,
                        'lon': lon}, ignore_index=True)
```

```python
location_key
```

```python
for entry in company_ids: 
    name = bols.loc[bols['company_location_id'] == entry]['company'].tolist()[0]
    address = bols.loc[bols['company_location_id'] == entry]['company_address'].tolist()[0]
    lat = bols.loc[bols['company_location_id'] == entry]['company_lat'].tolist()[0]
    lon = bols.loc[bols['company_location_id'] == entry]['company_lon'].tolist()[0]
    location_key  = location_key.append({'id': entry, 
                        'name': name,
                        'address': address,
                        'lat': lat,
                        'lon': lon}, ignore_index=True)
```
