### Innodata Viz

This visualization is attempting to show a relationship between shipments from Innodata to IA and the books scanned at Innodata per month. From interview with internet archive staff and videos on their website, I know that IA sends Innodata books to be digitized. After their digitization, Innodata sends the books back to the US. This chart is attempting to see if there are any spikes in books scanned in the time before the shipments (i.e., books returned to the US). 


````python

scans = alt.Chart("https://raw.githubusercontent.com/ers6/ia_bols/main/cebu_scan_counts.csv").mark_area(
    color='lightblue', 
    interpolate='step-after', 
    line=True).encode(
    x=alt.X('yearmonth(month_year):T', axis=alt.Axis(title="Months")),
    y=alt.Y('scandate:Q', axis=alt.Axis(title="Books Scanned"))
)
shipments = alt.Chart(bols.loc[bols['supplier']=='Innodata Knowledge Services Inc']).mark_rule(color='black').encode(
    x='arrival_date:T'
)

innodata_chart = alt.layer(
    scans, shipments
).properties(
    width=800, height=300
)
````

## Dashboarding 
trying to make a sample dashboard based on scan center location

````python
#  making bar chart and map 

from vega_datasets import data

selector=alt.selection_single(fields=['long', 'lat'], on='click', nearest=True)

# import point data
scan_center_locs = pd.read_csv("https://raw.githubusercontent.com/scanninglabor/IAScanningLabor/main/code/scans_per_center_per_year.csv")

# globe background

world = alt.topo_feature(data.world_110m.url, 'countries')

# US states background
background = alt.Chart(world).mark_geoshape(
    fill='lightgray',
    stroke='white'
).properties(
    width=800,
    height=600
).project('equalEarth')


# scan center locations

points = alt.Chart(scan_center_locs).mark_circle(
    size=20,
    color='steelblue'
).encode(
    longitude='long:Q',
    latitude='lat:Q', 
    tooltip=['name']
).add_selection(selector)

this_map = background + points


# making a bar chart for scans over time 

scans = alt.Chart(scan_center_locs).mark_area(
    color='lightblue', 
    interpolate='step-after', 
    line=True).encode(
    x=alt.X('date:T', axis=alt.Axis(title="Months")),
    y=alt.Y('count:Q', axis=alt.Axis(title="Books Scanned"))
).transform_filter(selector)


````

````python
# putting the bar chart and the map together into a dash
from vega_datasets import data

selector=alt.selection_single(fields=['long', 'lat'], on='click', nearest=True)

# import point data
scan_center_url = "https://raw.githubusercontent.com/scanninglabor/IAScanningLabor/main/code/scans_per_center_per_year.csv"
# globe background

world = alt.topo_feature(data.world_110m.url, 'countries')

# US states background
background = alt.Chart(world).mark_geoshape(
    fill='lightgray',
    stroke='white'
).properties(
    width=800,
    height=600
).project('equalEarth')


# scan center locations

points = alt.Chart(scan_center_url).mark_circle(
    size=20,
    color='steelblue'
).encode(
    longitude='long:Q',
    latitude='lat:Q', 
    tooltip=['name:N']
).add_selection(selector)

this_map = background + points


# making a bar chart for scans over time 

scans = alt.Chart(scan_center_url).mark_area(
    color='lightblue', 
    interpolate='step-after', 
    line=True).encode(
    x=alt.X('date:T', axis=alt.Axis(title="Months")),
    y=alt.Y('count:Q', axis=alt.Axis(title="Books Scanned"))
).transform_filter(selector).interactive()

draft_dash = this_map | scans
````
